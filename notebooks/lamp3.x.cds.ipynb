```python
DATA='../data/'
```

```python
import csv
best_matches = list(csv.DictReader(open(DATA + 'best_matches.tsv'), delimiter='\t'))
total_matches = list(csv.DictReader(open(DATA + 'total_matches.tsv'), delimiter='\t'))
```

```python
best_matches_p = [ float(b['percent matched letters']) for b in best_matches ]
```

```python
hist(best_matches_p, bins=20, normed=True, cumulative=True)
title('Cumulative dist of percent matched letters, for all matches')
xlabel('Fraction of sequences that match at that percent or lower')
```

```python
total_matches_p = [ float(b['percent matched letters']) for b in total_matches ]
```

```python
hist(total_matches_p, bins=20, cumulative=True, normed=True)
title('Cumulative dist of percent matched letters, for all matches')
xlabel('Fraction of sequences that match at that percent or lower')
```

```python
best_matches = list(sorted(best_matches, key=lambda x:x['percent matched letters']))
```
