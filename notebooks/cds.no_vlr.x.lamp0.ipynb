```python
%pylab inline
DATA='../data/'
```

```python
import csv
best_matches = csv.DictReader(open(DATA + 'p_marinus.no_vlr.x.lamp0.best_matches.tsv'), delimiter='\t')
total_matches = csv.DictReader(open(DATA + 'p_marinus.no_vlr.x.lamp0.total_matches.tsv'), delimiter='\t')
```

```python
best_matches = [ float(b['percent matched letters']) for b in best_matches ]
```

```python
hist(best_matches, bins=20, normed=True, cumulative=True)
title('Cumulative dist of percent matched letters, for all matches')
xlabel('Fraction of sequences that match at that percent or lower')
```

```python
total_matches = [ float(b['percent matched letters']) for b in total_matches ]
```

```python
hist(total_matches, bins=20, cumulative=True, normed=True)
title('Cumulative dist of percent matched letters, for all matches')
xlabel('Fraction of sequences that match at that percent or lower')
```
